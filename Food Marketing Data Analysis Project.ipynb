```python
# Ordering values in specific column in ascending order to check the age range in the DataFrame

food_marketing_data["Age"].sort_values(ascending = True)
```

```python
# Creating age groups - Placing a specific age to a certain age range category

age_groups = [(23,30),(31,40),(41,50),(51,60),(61,70),(71,85)]

def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown")

food_marketing_data["Age_Group"] = food_marketing_data["Age"].apply(assign_age_group)
```

#### OVERALL FINDINGS: <br>

##### 1. Age: People aged 30 to 70 spend more money, though they are less likely to accept campaigns. 
##### Younger groups tend to spend less and are less likely to engage with campaigns.<br>

##### 2. Catalog vs. Store/Web: Catalog customers are more likely to accept campaigns, but in-person shoppers spend more. 
##### A recommended strategy is to split focus across all channels: 40% Catalog, 30% Store, and 30% Web. <br>

##### 3. Children: People with no or fewer children tend to spend more. Focus efforts on this group. <br>

##### 4. Education: Education level does not seem to impact spending or campaign acceptance. Avoid targeting based on education. <br>

##### 5. Marital Status: While marital status doesn't significantly affect spending, Married, Single, and Together individuals spend more money. <br>


#### RECOMMENDATIONS: <br>
##### 1. Target Middle-Aged, High Earners with No Kids: Focus on people in the 30-70 age range with no children, as they tend to spend more. 
##### Use a split marketing approach across all platforms (40% Catalog, 30% Store, 30% Web). <br>
##### 2. Engage Younger and Older Age Groups (23-30 and 70+): These groups have higher campaign acceptance rates. 
##### Target them to drive new users and increase spending.
