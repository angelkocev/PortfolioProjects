#### Data Cleaning

```python
# Only want USA races, 50k or 50mi, 2020

fifty_df = df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020)]
fifty_df
```

```python
usa_df = fifty_df[fifty_df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA"]

usa_df.head(5)
```

```python
usa_df.shape
```

```python
usa_df.describe
```

```python
# Remove (USA) from Event Name column

usa_df["Event name"] = usa_df["Event name"].str.split("(").str.get(0)

usa_df.head(5)
```

```python
# clean up athlete age

usa_df["age"] = usa_df["Year of event"] - usa_df["Athlete year of birth"]

usa_df.head(5)
```

```python
# remove "h" from Athlete performance column

usa_df["Athlete performance"] = usa_df["Athlete performance"].str.split(" ").str.get(0)

usa_df.head(5)
```

```python
# Drop columns: Athlete club, Athlete year of birth, Athlete age category

usa_df = usa_df.drop(["Athlete club","Athlete year of birth","Athlete age category"], axis = 1)
usa_df.head(5)
```

```python
# Identifying Null values

usa_df.isna().sum()
```

```python
# Identifying Null values

usa_df[usa_df["age"].isna() == 1]
```

```python
# Remove Null values

usa_df = usa_df.dropna()
usa_df
```

```python
# Identifying Null values

usa_df.isna().sum()
```

```python
usa_df.shape
```

```python
# Identifying Duplicate Values

usa_df[usa_df.duplicated() == True]
```

```python
# Reseting Index

usa_df = usa_df.reset_index(drop = True)
```

```python
# Identifying data types

usa_df.dtypes
```

```python
# Fixing data types

usa_df["age"] = usa_df["age"].astype(int)
```

```python
# Fixing data types

usa_df["Athlete average speed"] = usa_df["Athlete average speed"].astype(float)
```

```python
usa_df.dtypes
```

```python
usa_df.head(5)
```

```python
# Renaming Columns

usa_ultra_marathon = usa_df.rename(columns = {
"Year of event":"year",
"Event dates":"race_day",
"Event name":"race_name",
"Event distance/length":"race_length",
"Event number of finishers":"number_of_finishers",
"Athlete performance":"athlete_performance",
"Athlete country":"athlete_country",
"Athlete gender":"athlete_gender",
"Athlete average speed":"athlete_average_speed",
"Athlete ID":"athlete_id"})
```

```python
usa_ultra_marathon.head(5)
```

```python
# Reordering Columns

ultra_marathon_usa = usa_ultra_marathon[["athlete_id","race_day","race_name","race_length","number_of_finishers", 
                                         "athlete_gender","athlete_country","athlete_average_speed","athlete_performance","age"]]
```

#### Exploratory Data Analysis

###### Questions that need to be answered:

1. Difference in speed for the 50km/50mi for male and female
2. What age groups are the best in the 50mi race (20+ races minimum)?
3. What age groups are the worst in the 50mi race (20+ races minimum)?
4. Are runners slower in summer than winter?

```python
# 1. Difference in speed for the 50km/50mi for male and female

ultra_marathon_usa.groupby(["race_length","athlete_gender"])["athlete_average_speed"].mean()
```

```python
# 2. What age groups are the best in the 50mi race (20+ races minimum)? (show 15)

ultra_marathon_usa.query("race_length == '50mi'").groupby("age")["athlete_average_speed"].agg(["mean","count"]).sort_values("mean", 
                                                                                                ascending = False).query("count > 19").head(15)
```

```python
# 3. What age groups are the worst in the 50mi race (20+ races minimum)? (show 15)

ultra_marathon_usa.query("race_length == '50mi'").groupby("age")["athlete_average_speed"].agg(["mean","count"]).sort_values("mean", 
                                                                                                ascending = True).query("count > 19").head(15)
```

```python
# 4. Are runners slower in summer than winter? (split between two decimals)

ultra_marathon_usa["race_month"] = ultra_marathon_usa["race_day"].str.split(".").str.get(1).astype(int)
ultra_marathon_usa.head(5)
```

```python
ultra_marathon_usa["race_season"] = ultra_marathon_usa["race_month"].apply(lambda x: 
                                                                           "Winter" if x > 11 
                                                                           else "Fall" if x > 8
                                                                           else "Summer" if x > 5
                                                                           else "Spring" if x > 2
                                                                           else "Winter")
```

```python
ultra_marathon_usa.head(5)
```

```python
# 4. Are runners slower in summer than winter? (split between two decimals)

ultra_marathon_usa.groupby("race_season")["athlete_average_speed"].agg(["mean","count"]).sort_values("mean", ascending = False)
```

```python
# 4. Are runners slower in summer than winter? (split between two decimals) (50 miler only)

ultra_marathon_usa.query("race_length == '50mi'").groupby("race_season")["athlete_average_speed"].agg(["mean","count"]).sort_values("mean", 
                                                                                                                                  ascending = False)
```
